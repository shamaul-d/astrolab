```python
import numpy as np
import matplotlib.pyplot as plt
import csv

results = []
with open("SiC-All.csv") as csvfile:
    reader = csv.reader(csvfile)
    for row in reader: # each row is a list
        results.append(row)

for i,row in enumerate(results):
    results[i] = row[11], row[13]

data = []
for i,r in enumerate(results[1:]):
    data.append(r)

reals = []

for i in data:
    if i[0] != '' and i[1] != '' and i[0] != '12C/13C':
        reals.append((i[0], i[1]))

#print(reals)

d = [0] * len(reals)

for j, row in enumerate(reals):
    if row[0] != '12C/13C':
        d[j] = [float(row[0]), float(row[1])]
    
#print(d)

n_data = np.array(d)
#print(n_data)
```

```python
results = []
with open("P040z14S4_abunds.csv") as csvfile:
    reader = csv.reader(csvfile)
    for row in reader: # each row is a list
        results.append(row)

t = [0] * len(results)        
        
for i,row in enumerate(results):
    results[i] = row[5:12]
    t[i] = row[12]

data1 = results[1:]

for j, row in enumerate(data1):
    data1[j] = [float(i) for i in row]

#title = results[0::8]
#C12 = data[0:]
#C13 = data[1:]
#N14 = data[2:]
#N15 = data[3:]
#O16 = data[4:]
#O17 = data[5:]
#O18 = data[6:]

n_data1 = np.array(data1)
#print(n_data.shape)

C12 = n_data1[:,0]
C13 = n_data1[:,1]
N14 = n_data1[:,2]
N15 = n_data1[:,3]
O16 = n_data1[:,4]
O17 = n_data1[:,5]
O18 = n_data1[:,6]

c12c13o = C12[:19958:50] / C13[:19958:50]
n14n15o =  N14[:19958:50] / N15[:19958:50]
o17o16o = O17[:19958:50] / O16[:19958:50]
o18o16o = O18[:19958:50] / O16[:19958:50]

c12c13bs = C12[19958:26050:50] / C13[19958:26050:50]
n14n15bs =  N14[19958:26050:50] / N15[19958:26050:50]
o17o16bs = O17[19958:26050:50] / O16[19958:26050:50]
o18o16bs = O18[19958:26050:50] / O16[19958:26050:50]

c12c13ys = C12[20650:22023:50] / C13[20650:22023:50]
n14n15ys =  N14[20650:22023:50] / N15[20650:22023:50]
o17o16ys = O17[20650:22023:50] / O16[20650:22023:50]
o18o16ys = O18[20650:22023:50] / O16[20650:22023:50]

c12c13bs2 = C12[22023:24586:50] / C13[22023:24586:50]
n14n15bs2 =  N14[22023:24586:50] / N15[22023:24586:50]
o17o16bs2 = O17[22023:24586:50] / O16[22023:24586:50]
o18o16bs2 = O18[22023:24586:50] / O16[22023:24586:50]

c12c13wnl = C12[24586:73916:50] / C13[24586:73916:50]
n14n15wnl =  N14[24586:73916:50] / N15[24586:73916:50]
o17o16wnl = O17[24586:73916:50] / O16[24586:73916:50]
o18o16wnl = O18[24586:73916:50] / O16[24586:73916:50]

c12c13wc = C12[73916:83510:50] / C13[73916:83510:50]
n14n15wc =  N14[73916:83510:50] / N15[73916:83510:50]
o17o16wc = O17[73916:83510:50] / O16[73916:83510:50]
o18o16wc = O18[73916:83510:50] / O16[73916:83510:50]

c12c13wo = C12[83510::50] / C13[83510::50]
n14n15wo =  N14[83510::50] / N15[83510::50]
o17o16wo = O17[83510::50] / O16[83510::50]
o18o16wo = O18[83510::50] / O16[83510::50]

#print(c12c13)
#print(c12c13[:,0])
```
