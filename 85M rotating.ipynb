```python
import numpy as np
import matplotlib.pyplot as plt
import csv

results = []
with open("SiC-All.csv") as csvfile:
    reader = csv.reader(csvfile)
    for row in reader: # each row is a list
        results.append(row)

for i,row in enumerate(results):
    results[i] = row[11], row[13]

data = []
for i,r in enumerate(results[1:]):
    data.append(r)

reals = []

for i in data:
    if i[0] != '' and i[1] != '' and i[0] != '12C/13C':
        reals.append((i[0], i[1]))

#print(reals)

d = [0] * len(reals)

for j, row in enumerate(reals):
    if row[0] != '12C/13C':
        d[j] = [float(row[0]), float(row[1])]
    
#print(d)

n_data = np.array(d)
#print(n_data)
```
